```python
def change(lst):
  lst.append(99)

data = [1, 2]
change(data)
print(data)
```

Reflection Questions:

. Was the object mutable or immutable? The object is mutable

. Did Python mutate the object or rebind the name? The function mutated the existing object

. Did the change affect other references? Yes, since data and lst point to the same object, the change is visible globally

```python
def reset_wrong(lst):
  lst = []

def reset_right(lst):
  lst.clear()

data = [1,2,3]
reset_wrong(data)
print(data)

reset_right(data)
print(data)
```

Reflection Questions:

. Was the object mutable or immutable? The object is mutable

. Did Python mutate the object or rebind the name? In reset_wrong, Python rebound the name lst.
In reset_right, Python mutated the object using the .clear() method

. Did the change affect other references? The rebinding in reset_wrong did not affect other references.
The mutation in reset_right did affect all other references to that list

```python
def demo(x):
  print(id(x))
  x.append(5)
  print(id(x))

lst = [1,2]
print(id(lst))
demo(lst)
```

```python
a = [1,2]
b = [1,2]

print(a == b)
print(a is b)
```

Reflection Questions:

. Was the object mutable or immutable? The objects are mutable

. Did Python mutate the object or rebind the name? Neither occurred in the print statements; however, the initial assignments bound the names a and b to two separate list objects that happen to have the same values

. Did the change affect other references? Since they are distinct objects in memory (a is b is False), changing list a would not affect list b
