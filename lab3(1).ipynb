```python
#Part 3 - Generator: Lazy  Data Production
def sensor_stream(data):
    for value in data:
     print("Producing:", value)
     yield value
```

```python
stream = sensor_stream(sensor_readings)
next(stream)
```

```python
#Part 4 — yield vs return
def normal_func():
  print("Start")
  return 1
  print("End")

normal_func()
```

```python
def gen_func():
  print("Start")
  yield 1
  print("Middle")
  yield 2
  print("End")
  yield 3

g = gen_func()

next(g)
next(g)
next(g)

```

```python
#Part 5 — Generator State Is Preserved
def counter():
  x = 0
  while True:
    yield x
    x += 1

c = counter()
next(c)
next(c)
next(c)
```

```python
#Part 6 — Generator Pipeline (Controlled Flow)
def high_temp_filter(stream, threshold):
  for value in stream:
    print("Filtering:", value)
    if value > threshold:
      yield value
```

```python
stream = sensor_stream(sensor_readings)
filtered = high_temp_filter(stream, 23.0)
next(filtered)
next(filtered)
```

```python
#Part 7 — Dunder (Magic) Methods
class User:
  def __init__(self, name, age):
    self.name = name
    self.age = age
    def __str__(self):
      return f"{self.name}, age {self.age}"
      def __repr__(self):
        return f"User(name={self.name!r}, age={self.age})"

u = User("Alice", 21)
print(u)
u
```

```python
#7.2 Truth and Size
#__len__ , __bool__
class Cart:
  def __init__(self, items):
    self.items = items

  def __len__(self):
    return len(self.items)

  def __bool__(self):
    return len(self.items) > 0

cart = Cart(["apple", "banana"])
empty_cart = Cart([])

len(cart)
bool(cart)

if cart:
  print("Cart has items")
```

```python
#7.3 Indexing and Iteration
#__getitem__ , __iter__ , __next__
class Log:
  def __init__(self, data):
    self.data = data

  def __getitem__(self, index):
    return self.data[index]

log = Log([10, 20, 30, 40])

log[0]
log[2]
for x in log:
  print(x)
```

```python
#7.4 Comparison and Operators
#__eq__ , __lt__ , __add__
class Score:
  def __init__(self, value):
    self.value = value

  def __eq__(self, other):
    return self.value == other.value

  def __lt__(self, other):
    return self.value < other.value

  def __add__(self, other):
    return Score(self.value + other.value)

a = Score(10)
b = Score(20)

a == b
a < b

c = a + b
c.value
```
